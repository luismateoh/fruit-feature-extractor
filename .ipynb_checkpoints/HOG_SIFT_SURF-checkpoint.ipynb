```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('database/FRESA/fresa (39).jpg')
# gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

winSize = (64, 64)
blockSize = (16, 16)
blockStride = (8, 8)
cellSize = (8, 8)
nbins = 9
hog = cv.HOGDescriptor(winSize, blockSize, blockStride, cellSize, nbins)

descriptor = hog.compute(img).T
keypoints = []
for i in range(descriptor.shape[1]):
    pt = (int(descriptor[0][i]), int(descriptor[1][i]))
    keypoints.append(cv.KeyPoint(pt, 1))

# Draw the keypoints on the image
img_with_keypoints = cv.drawKeypoints(img, keypoints, None, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Display the image with the keypoints
plt.figure(figsize=(10, 10))
plt.imshow(img_with_keypoints)
plt.axis('off')
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('database/FRESA/fresa (39).jpg')

# Compute HOG descriptor
winSize = (64, 64)
blockSize = (16, 16)
blockStride = (8, 8)
cellSize = (8, 8)
nbins = 9
hog = cv.HOGDescriptor(winSize, blockSize, blockStride, cellSize, nbins)

# Compute the HOG descriptor
descriptor = hog.compute(img)

# Reshape the descriptor array and extract (x, y) coordinates
descriptor = descriptor.reshape(-1, nbins)
x = descriptor[:, 0]
y = descriptor[:, 1]

# Create keypoints from the (x, y) coordinates
keypoints = [cv.KeyPoint(pt[0], pt[1], 1) for pt in zip(x, y)]

# Draw the keypoints on the image
img_with_keypoints = cv.drawKeypoints(img, keypoints, None, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Display the image with the keypoints
plt.figure(figsize=(10, 10))
plt.imshow(img_with_keypoints)
plt.axis('off')
plt.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('database/FRESA/fresa (39).jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
sift = cv.SIFT_create()
kp = sift.detect(gray, None)
img = cv.drawKeypoints(gray, kp, img)
cv.imwrite('sift_keypoints.jpg', img)

img = cv.drawKeypoints(gray, kp, img, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
cv.imwrite('sift_keypoints.jpg', img)

plt.figure(figsize=(10, 10))
plt.imshow(img), plt.title('SIFT')
plt.axis('off')

plt.show()
```
